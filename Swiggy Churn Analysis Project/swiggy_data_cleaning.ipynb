```python
df=df.dropna(subset=['value'])
```

Fill null values


```python
df['value']=df['value'].fillna(0)
df['metric_name']=df['metric_name'].fillna('Unknown')
med=df['is_estimate'].median()
df['is_estimate']=df['is_estimate'].fillna(med)
```

Fill null date /time series data

```python
df['year_start']=df['year_start'].ffill() #forward fill
df['year_end']=df['year_end'].bfill() #backward fill 
```

```python
df[df['value'].isnull()]
```

CHECK DUPLICATES AND REMOVE IT

```python
print(df.duplicated().sum())
df[df.duplicated(keep=False)] #view duplicate rows
df=df.drop_duplicates() # drop exact duplicates and keep first occurance
df=df.sort_values('year_start',ascending=False)
```

FIXING DATA TYPES

```python
df['value'] = df['value'].astype(str).str.replace(',', '', regex=False).str.strip().astype(float)
```

STRING CLEANING

```python
df['metric_name']= df['metric_name'].str.strip()
df['sector'] =df['sector'].str.strip()
df['country']=df['country'].str.strip()
```

STANDARD CAPTILIZATION

```python
df['metric_name']=df['metric_name'].str.title()
df['metric_category']=df['metric_category'].str.title()
```

```python
Year_map={'FY2020':'2020','FY2021':'2021','FY2022':'2022','FY2023':'2023','FY2024':'2024','FY2025':'2025','Q1 FY2026':'2026'}
df['year_lable']=df['year_label'].replace(Year_map)
```

```python
print(df['year_lable'].value_counts())#checks
```

EXPORT THE CLEANED DATA

```python
df.to_excel(r'M:\Resume\churn project\Swiggy_cleaned_data.xlsx', index=False)
```
